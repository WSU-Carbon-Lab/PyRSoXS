## Parameter List

Before running the code, please input the following parameters:

- `En`: Photon Energy in eV (float)
- `d1`: Distance from axis of rotation to sample (mm) (float)
- `d2`: Sample - Detector distance (mm) (float)
- `beamCenterY`: y pixel value of the beam center (int)
- `beamCenterX`: x pixel value of the beam center (int)
- `phi`: Angle the CCD is rotated to (Input: degrees) (int)
- `PixelSize`: Size of a pixel (mm) (int)
- `num_pixels`: Number of pixels along one dimension of the image (int)
- `bins`: Number of bins used for Histogram (int)

Please replace the placeholder values in the code with your actual parameter values before executing it.

Example:

```python
    # Assign values to parameters
    params1 = RGC.Parameters(En = 320, phi = 17.5, d1 = 26, d2 = 23, beamCenterY = 556, beamCenterX = 529, PixelSize = 0.027, num_pixels = 1024, bins = 200)
```
